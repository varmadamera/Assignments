```python
zero_indexes = prisoners.loc[prisoners['No. of Inmates benefitted by Elementary Education']==0]
display  (zero_indexes)
```

```python
prisoners["Total_Benefited"]=prisoners.iloc[:,2:].sum(axis=1)
display (prisoners)
```

```python
prisoners_total  =  prisoners
display (prisoners_total)
```

```python
prisoners_total['YEAR'] =prisoners_total['YEAR'].astype('str')
```

```python
prisoners_total.loc['Prisoners Total']= prisoners_total.iloc[:, 2:].sum(numeric_only=True)
display (prisoners_total)
```

```python
prisoners_total.loc[prisoners_total.index[-1], ['STATE/UT','YEAR']] = ''
display (prisoners_total)
```

```python
import numpy as np
xlabels = prisoners['STATE/UT'].values
plot.figure(figsize=(20, 10))

plot.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical', fontsize=18)
plot.xticks

plot.bar(np.arange(prisoners.iloc[:-1].values.shape[0]),prisoners['Total_Benefited'].iloc[:-1],align='edge')
plot.show()
```

```python
labels = prisoners.columns[2:-1]
sizes = prisoners_total.iloc[-1][2:-1].values
fig1, ax1 = plot.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plot.show()
```
